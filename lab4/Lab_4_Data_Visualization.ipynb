# Data Visualization

## Histogram
* A histogram of the salaries for women
* Add the men's salaries to this histogram
* Give your histogram a dark background and label the axes
* Next, please label the colors of the histogram (men vs women)

```python
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt

df = pd.read_csv("data/school_earnings.csv")
#df = df.sort_values(by=['Women'], ascending=False)
print(df.head(5))
```

```python
women = df.loc[:, 'Women'].values
men = df.loc[:, 'Men'].values
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.set_facecolor('gray')
plt.hist(women, bins=10, color='Pink', label='Women')
plt.hist(men, bins=10, color='SkyBlue', alpha=0.5, label='Men')
plt.legend(loc='upper right')
plt.show()
```

## Bar chart
* Instead of a histogram, create a bar chart that lists the salary for women (y-axis) for each school
* Also add men to this bar chart
* Make the style nice and add labels

```python
import numpy as np

index = df.loc[:, 'School']
index_num = np.arange(len(women))
fig, ax = plt.subplots()

w = ax.bar(index_num, women, 0.35, 0.6, color='Pink', label='Women')
m = ax.bar(index_num+0.35, men, 0.35, 0.6, color='SkyBlue', label='Men')
plt.xticks(index_num, index, rotation=90)
plt.legend(loc='upper right')
plt.show()
```

## Box plot
Now create a nice boxplot of the data, for each sex, on the same graph

```python
bps = plt.boxplot([women, men], widths=0.3, labels=['Women','Men'], patch_artist=True)
colors = ['pink', 'skyblue']
for i in range(len(bps['boxes'])):
    bps['boxes'][i].set_facecolor(colors[i])
plt.show()
```
